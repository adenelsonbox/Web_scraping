```python
page = 1
dict1 = {"Name": [],
          "oldPrice": [],
          "newPrice" : [],
          "Discount" : []
        }
while page <= 50:
    for item in soup.find_all("a", class_ = "core"):
        try:
            dict1["Name"].append(item.find("h3",class_="name").text)
        except:
            dict1["Name"].append(np.nan)
        try:
            dict1["oldPrice"].append(item.find("div",class_="old").text)
        except:
            dict1["oldPrice"].append(np.nan)
        try:
            dict1["newPrice"].append(item.find("div",class_="prc").text)
        except:
            dict1["newPrice"].append(np.nan)
        try:
            dict1["Discount"].append(item.find("div", class_="bdg _dsct _sm").text)
        except:
            dict1["Discount"].append(np.nan)
    page+=1
```

```python
df = pd.DataFrame(dict1)
```
